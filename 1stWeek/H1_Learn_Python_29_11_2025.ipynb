```python
# ==========================================
# BAGIAN 1: PERSIAPAN (Import Library)
# ==========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Agar grafik terlihat bagus di Tablet
sns.set_style("whitegrid")

# ==========================================
# BAGIAN 2: MEMBUAT DATA CONTOH (DUMMY)
# Nanti bagian ini dihapus jika pakai data asli RS Anda
# ==========================================
data = {
    'No_RM': [f'RM{i:04d}' for i in range(1, 101)],
    'Tanggal_Kunjungan': pd.date_range(start='2024-01-01', periods=100),
    'Umur': np.random.randint(5, 85, 100), # Umur acak 5-85 tahun
    'Jenis_Kelamin': np.random.choice(['Laki-laki', 'Perempuan'], 100),
    'Nama_Poli': np.random.choice(['Poli Katarak', 'Poli Retina', 'Poli Glaukoma', 'Poli Infeksi'], 100, p=[0.4, 0.2, 0.1, 0.3]),
    'Kecamatan': np.random.choice(['Makassar', 'Ujung Pandang', 'Panakkukang', 'Tamalanrea'], 100)
}
df = pd.DataFrame(data)

# -----------------------------------------------------------
# CARA PAKAI DATA ASLI ANDA (EXCEL/CSV):
# 1. Upload file Excel Anda ke menu folder di kiri Colab.
# 2. Hapus kode BAGIAN 2 di atas, ganti dengan:
# df = pd.read_excel('NamaFileAnda.xlsx')
# -----------------------------------------------------------

# ==========================================
# BAGIAN 3: ANALISIS DATA (Hanya 1 Klik!)
# ==========================================

print("=== 5 Data Teratas ===")
display(df.head())

# 1. Analisis Umur Pasien (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Umur', kde=True, color='skyblue')
plt.title('Distribusi Umur Pasien RS Mata', fontsize=15)
plt.xlabel('Umur (Tahun)')
plt.ylabel('Jumlah Pasien')
plt.show()

# 2. Analisis Jumlah Pasien per Poli (Bar Chart)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Nama_Poli', order=df['Nama_Poli'].value_counts().index, palette='viridis')
plt.title('Poli Paling Sibuk (Beban Kerja)', fontsize=15)
plt.xlabel('Jumlah Kunjungan')
plt.ylabel('Nama Poli')
plt.show()

# 3. Analisis Umur Rata-rata per Poli (Boxplot)
# Ini menjawab: "Apakah pasien retina lebih muda dari katarak?"
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Umur', y='Nama_Poli', palette='coolwarm')
plt.title('Sebaran Umur Pasien Berdasarkan Poli', fontsize=15)
plt.show()

# Tambahkan parameter hue='Nama_Poli' dan legend=False agar warning hilang
sns.boxplot(data=df, x='Nama_Poli', y='Umur', hue='Nama_Poli', palette='coolwarm')
plt.title('Sebaran Umur Pasien Berdasarkan Poli', fontsize=15)
plt.show()
```
